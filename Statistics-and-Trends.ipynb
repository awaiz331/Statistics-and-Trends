## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

## to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

## Reading Dataset

```python
df = pd.read_csv('final_demographics_data.csv',index_col='country')
df.head(10)
```

```python
df.describe()
```

## Checking NaN values

```python
# Checking column wise nan values
nan_counts = df.isna().sum()
print(nan_counts)
```

## Replacing NaN with their column average (Pre-Procesing)

```python
# Replace NaN values in numerical columns with the column mean
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    df[column].fillna(df[column].mean(), inplace=True)

# Replace NaN values in categorical columns with the mode
for column in df.select_dtypes(include=['object']).columns:
    if df[column].isna().sum() > 0:
        df[column].fillna(df[column].mode()[0], inplace=True)

df.head(5)
```

## Correlation Matrix

```python
df1 = df.drop(df.columns[0], axis=1) ## droping col which has non numeric values
df1.corr()
```

## Summary

```python

# Calculate the major moments
skewness = df1.skew()
kurt = df1.kurt()

# Combine all statistics into a single DataFrame
summary_df = pd.DataFrame({
    'Skewness': skewness,
    'Kurtosis': kurt
})

# Print the summary DataFrame
print("\nSummary Statistics:")
print(summary_df)
```

# Plotting

### Heatmap

```python
import seaborn as sns

def plot_heatmap(df):
    """
    Plots a heatmap of the correlation matrix to show relationships between numerical variables.
    
    Parameters:
        df (DataFrame): The dataset.
    """
    plt.figure(figsize=(12, 8))
    sns.heatmap(df.corr(), annot=False, cmap='coolwarm', linewidths=0.5)
    plt.title('Heatmap of Correlation Matrix')
    plt.show()

# Example usage
plot_heatmap(df1)
```

## Pie Chart

```python
def plot_pie(df, column):
    """
    Plots a pie chart to show the distribution of categorical data.
    
    Parameters:
        df (DataFrame): The dataset.
        column (str): Column name for the pie chart.
    """
    plt.figure(figsize=(8, 8))
    df[column].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title(f'Pie Chart of {column}')
    plt.ylabel('')
    plt.show()

# Example usage
plot_pie(df, 'Region')
```

### Bar Chart

```python
def bar_chart(df):
    """
    Generates a bar chart comparing average military expenditure by region.

    Parameters:
    dataframe : df
    
    - 'Region': A categorical column indicating the region.
    - 'Military expenditure (% of GDP)': A numerical column indicating military expenditure as a percentage of GDP.

    Returns:
    None: Displays a bar chart of average military expenditure by region.
    """
    # --- Plot: Bar Chart for Categorical Comparison ---
    plt.figure(figsize=(12, 8))
    # Aggregate data by region and calculate the mean military expenditure
    region_military_expenditure = df.groupby('Region')['Military expenditure (% of GDP)'].mean().sort_values()
    region_military_expenditure.plot(kind='bar', color='skyblue')
    plt.title('Average Military Expenditure by Region (2020)', fontsize=14)
    plt.xlabel('Region', fontsize=12)
    plt.ylabel('Military Expenditure (% of GDP)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Example usage
bar_chart(df)
```

### Scatter Plot

```python
def scatter_plot(df, x_col, y_col, title, x_label, y_label):
    """
    Plots a scatter plot to analyze the relationship between two continuous variables.

    Parameters:
        df (DataFrame): The dataset.
        x_col (str): Column name for the x-axis.
        y_col (str): Column name for the y-axis.
        title (str): Title for the plot.
        x_label (str): Label for the x-axis.
        y_label (str): Label for the y-axis.
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df[x_col], y=df[y_col], alpha=0.7)
    plt.title(title, fontsize=14)
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.grid(visible=True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Example usage
scatter_plot(df, 
             'GDP per capita (constant 2010 US$)', 
             'Unemployment, total (% of total labor force) (modeled ILO estimate)', 
             'GDP per Capita vs Unemployment Rate (2020)', 
             'GDP per Capita (constant 2010 US$)', 
             'Unemployment Rate (%)')
```

## Plot Box

```python
def box_plot():
    """
    Plots a box plot to show the distribution of a continuous variable and detect outliers.
    
    Parameters:
        df (DataFrame): The dataset.
        return None and plopt the graph
    """
    
    plt.figure(figsize=(12, 8))
    sns.boxplot(
        x='Region',
        y='Age dependency ratio (% of working-age population)',
        data=df
    )
    plt.title('Age Dependency Ratio by Region (2020)', fontsize=14)
    plt.xlabel('Region', fontsize=12)
    plt.ylabel('Age Dependency Ratio (%)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

box_plot()
```

## Correlation Heatmap For Selected Columns

```python
def plot_correlation_heatmap(df, columns, title):
    """
    Calculates and plots a heatmap of the correlation matrix for selected columns.

    Parameters:
        df (DataFrame): The dataset.
        columns (list of str): List of column names to include in the correlation matrix.
        title (str): Title for the heatmap.
    """
    # Calculate the correlation matrix for the selected columns
    correlation_matrix = df[columns].corr()

    # Plot the heatmap for the correlation matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', 
                square=True, linewidths=0.5, cbar_kws={"shrink": .8})
    plt.title(title, fontsize=14)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Example usage
plot_correlation_heatmap(df, 
                         ['GDP per capita (constant 2010 US$)', 
                          'Unemployment, total (% of total labor force) (modeled ILO estimate)',
                          'Military expenditure (% of GDP)',
                          'Age dependency ratio (% of working-age population)'],
                         'Correlation Heatmap of Selected Economic Indicators')
```

### Pair Plots Between Selected Columns

```python
def plot_pairplot(df, columns, title):
    """
    Plots a pair plot to show pairwise relationships between selected continuous variables.
    
    Parameters:
        df (DataFrame): The dataset.
        columns (list of str): List of column names to include in the pair plot.
        title (str): Title for the plot.
    """
    plt.figure(figsize=(10, 8))
    sns.pairplot(df[columns])
    plt.suptitle(title, y=1.02)
    plt.show()

# Example usage
plot_pairplot(df, 
              ['GDP per capita (constant 2010 US$)', 
               'Unemployment, total (% of total labor force) (modeled ILO estimate)',
               'Military expenditure (% of GDP)',
               'Age dependency ratio (% of working-age population)'],
              'Pair Plot of Selected Economic and Demographic Indicators (2020)')
```
